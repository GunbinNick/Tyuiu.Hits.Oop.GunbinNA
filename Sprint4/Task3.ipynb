демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) создайте явную реализации интерфейса и управление зависимостями

```
using System;
using System.Collections.Generic;

public abstract class Product
{
    public string Name { get; set; }
    public string Manufacturer { get; set; }
    public int Price { get; set; }
    public double Weight { get; set; } // Добавлен атрибут
    public DateTime ReleaseDate { get; set; } // Добавлен атрибут

    public Product(string name, string manufacturer, int price, double weight, DateTime releaseDate)
    {
        Name = name;
        Manufacturer = manufacturer;
        Price = price;
        Weight = weight;
        ReleaseDate = releaseDate;
    }

    public virtual string GetInfo()
    {
        return $"Название - {Name}; Производитель - {Manufacturer}; Цена - {Price} руб.; Вес - {Weight} кг.; Дата выпуска - {ReleaseDate.ToShortDateString()}";
    }

    public virtual void Discount()
    {
        Console.WriteLine($"Применена скидка на '{Name}' от производителя {Manufacturer}. Итоговая стоимость - {Price * 0.75} руб.");
    }

    public virtual void Display()
    {
        Console.WriteLine($"Отображение товара {Name} от {Manufacturer} на экране.");
    }

    public virtual void CalculateShippingCost()
    {
        double shippingCost = Weight * 10; // Расчет стоимости доставки
        Console.WriteLine($"Стоимость доставки для {Name} - {shippingCost} руб.\n");
    }
}

public interface IClothes
{
    int Size { get; set; }
    void CheckFit(int userSize);
}

public class Clothing : Product, IClothes
{
    public int Size { get; set; }
    public string Material { get; set; } // Добавлен атрибут
    public string Color { get; set; } // Добавлен атрибут

    public Clothing(string name, string manufacturer, int price, double weight, DateTime releaseDate, int size, string material, string color) : base(name, manufacturer, price, weight, releaseDate)
    {
        Size = size;
        Material = material;
        Color = color;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Размер - {Size}; Материал - {Material}; Цвет - {Color}";
    }

    public override void Display()
    {
        Console.WriteLine($"Отображение товара {Name} от {Manufacturer} размера {Size} на экране.");
    }

    // Явная реализация интерфейса IClothes
    void IClothes.CheckFit(int userSize)
    {
        if (userSize == Size)
        {
            Console.WriteLine($"Одежда {Name} подходит по размеру.\n");
        }
        else
        {
            Console.WriteLine($"Одежда {Name} не подходит по размеру.\n");
        }
    }

    public void ChangeColor(string newColor)
    {
        Color = newColor;
        Console.WriteLine($"Цвет одежды {Name} изменен на {Color}.\n");
    }
}

public class Books : Product
{
    public string Author { get; set; }
    public int Pages { get; set; } // Добавлен атрибут
    public string ISBN { get; set; } // Добавлен атрибут

    public Books(string name, string manufacturer, int price, double weight, DateTime releaseDate, string author, int pages, string isbn) : base(name, manufacturer, price, weight, releaseDate)
    {
        Author = author;
        Pages = pages;
        ISBN = isbn;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Автор - {Author}; Страниц - {Pages}; ISBN - {ISBN}";
    }

    public void CalculateReadingTime(int readingSpeed)
    {
        double hours = Pages / readingSpeed;
        Console.WriteLine($"Время чтения книги '{Name}' - {hours} часов.\n");
    }

    public void CheckAvailability()
    {
        Console.WriteLine($"Книга '{Name}' доступна в наличии.\n");
    }
}

public class Fiction : Books
{
    public string Genre { get; set; }

    public Fiction(string name, string manufacturer, int price, double weight, DateTime releaseDate, string author, int pages, string isbn, string genre) : base(name, manufacturer, price, weight, releaseDate, author, pages, isbn)
    {
        Genre = genre;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Жанр - {Genre} ";
    }

    public void RecommendSimilarBooks()
    {
        Console.WriteLine($"Рекомендуем книги в жанре {Genre}.\n");
    }
}

public class Electronics : Product
{
    public int WarrantyPeriod { get; set; }
    public int PowerConsumption { get; set; } // Добавлен атрибут
    public string Model { get; set; } // Добавлен атрибут

    public Electronics(string name, string manufacturer, int price, double weight, DateTime releaseDate, int warrantyPeriod, int powerConsumption, string model) : base(name, manufacturer, price, weight, releaseDate)
    {
        WarrantyPeriod = warrantyPeriod;
        PowerConsumption = powerConsumption;
        Model = model;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Гарантия - {WarrantyPeriod} года; Потребление энергии - {PowerConsumption} Вт; Модель - {Model}";
    }

    public override void Discount()
    {
        Console.WriteLine($"Применена скидка на '{Name}' от производителя {Manufacturer} с гарантией {WarrantyPeriod} года. Итоговая стоимость - {Price * 0.75} руб., Гарантия переопределена - {WarrantyPeriod / 2} года");
    }

    public void CalculateEnergyCost(int hoursPerDay, int daysPerYear, double pricePerKWh)
    {
        double energyConsumption = PowerConsumption * hoursPerDay * daysPerYear / 1000.0;
        double totalCost = energyConsumption * pricePerKWh;
        Console.WriteLine($"Годовая стоимость энергии для {Name} - {totalCost} руб.\n");
    }

    public void CheckCompatibility(string otherModel)
    {
        Console.WriteLine($"Проверка совместимости {Name} с моделью {otherModel}.\n");
    }
}

// Пример вызова метода GetInfo и вывода результата
Fiction diction = new Fiction("1984", "Издательство АСТ", 239, 0.5, new DateTime(1949, 6, 8), "Дж. Оруэлл", 328, "978-5-17-103394-7", "Художественная литература");
Console.WriteLine(diction.GetInfo());
diction.CalculateReadingTime(20);
diction.CheckAvailability();
diction.RecommendSimilarBooks();

Clothing clother = new Clothing("Худи", "OffWhite", 19299, 0.8, new DateTime(2023, 1, 1), 56, "Хлопок", "Черный");
Console.WriteLine(clother.GetInfo());
clother.Display();
// Вызов явно реализованного метода интерфейса
((IClothes)clother).CheckFit(56);
clother.ChangeColor("Белый");

Electronics tv = new Electronics("Samsung G5120", "Samsung", 42999, 5.0, new DateTime(2022, 1, 1), 4, 150, "G5120");
Console.WriteLine(tv.GetInfo());
tv.Discount();
tv.Display();
tv.CalculateEnergyCost(5, 365, 0.1);
tv.CheckCompatibility("G5120");

Books mybook = new Books("Удержать небо", "Издательство АСТ", 1099, 0.6, new DateTime(2021, 1, 1), "Лю Цысинь", 448, "978-5-17-103394-7");
Console.WriteLine(mybook.GetInfo());
mybook.CalculateReadingTime(25);
mybook.CheckAvailability();

// Generic-класс для хранения списка продуктов
public class ProductList<T> where T : Product
{
    private List<T> products = new List<T>();

    public void AddProduct(T product)
    {
        products.Add(product);
    }

    public void DisplayAll()
    {
        foreach (var product in products)
        {
            product.Display();
        }
    }
}

ProductList<Product> productList = new ProductList<Product>();
productList.AddProduct(diction);
productList.AddProduct(clother);
productList.AddProduct(tv);
productList.AddProduct(mybook);

productList.DisplayAll();
```
