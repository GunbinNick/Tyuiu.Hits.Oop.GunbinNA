```
public abstract class Product
{
    public string Name { get; set; }
    public string Manufacturer { get; set; }
    public int Price { get; set; }
    public double Weight { get; set; } // Добавлен атрибут

    public Product(string name, string manufacturer, int price, double weight)
    {
        Name = name;
        Manufacturer = manufacturer;
        Price = price;
        Weight = weight;
    }

    public virtual string GetInfo()
    {
        return $"Название - {Name}; Производитель - {Manufacturer}; Цена - {Price} руб.; Вес - {Weight} кг.";
    }

    public virtual void Discount()
    {
        Console.WriteLine($"Применена скидка на '{Name}' от производителя {Manufacturer}. Итоговая стоимость - {Price * 0.75} руб.");
    }

    public virtual void Display()
    {
        Console.WriteLine($"Отображение товара {Name} от {Manufacturer} на экране.");
    }

    public virtual void CalculateShippingCost()
    {
        double shippingCost = Weight * 10; // Расчет стоимости доставки
        Console.WriteLine($"Стоимость доставки для {Name} - {shippingCost} руб.\n");
    }
}

public interface IClothes
{
    int Size { get; set; }
}

public class Clothing : Product, IClothes
{
    public int Size { get; set; }
    public string Material { get; set; } // Добавлен атрибут

    public Clothing(string name, string manufacturer, int price, double weight, int size, string material) : base(name, manufacturer, price, weight)
    {
        Size = size;
        Material = material;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Размер - {Size}; Материал - {Material}";
    }

    public override void Display()
    {
        Console.WriteLine($"Отображение товара {Name} от {Manufacturer} размера {Size} на экране.");
    }

    public void CheckFit(int userSize)
    {
        if (userSize == Size)
        {
            Console.WriteLine($"Одежда {Name} подходит по размеру.\n");
        }
        else
        {
            Console.WriteLine($"Одежда {Name} не подходит по размеру.\n");
        }
    }
}

public class Books : Product
{
    public string Author { get; set; }
    public int Pages { get; set; } // Добавлен атрибут

    public Books(string name, string manufacturer, int price, double weight, string author, int pages) : base(name, manufacturer, price, weight)
    {
        Author = author;
        Pages = pages;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Автор - {Author}; Страниц - {Pages}";
    }

    public void CalculateReadingTime(int readingSpeed)
    {
        double hours = Pages / readingSpeed;
        Console.WriteLine($"Время чтения книги '{Name}' - {hours} часов.\n");
    }
}

public class Fiction : Books
{
    public string Genre { get; set; }

    public Fiction(string name, string manufacturer, int price, double weight, string author, int pages, string genre) : base(name, manufacturer, price, weight, author, pages)
    {
        Genre = genre;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Жанр - {Genre} ";
    }
}

public class Electronics : Product
{
    public int WarrantyPeriod { get; set; }
    public int PowerConsumption { get; set; } // Добавлен атрибут

    public Electronics(string name, string manufacturer, int price, double weight, int warrantyPeriod, int powerConsumption) : base(name, manufacturer, price, weight)
    {
        WarrantyPeriod = warrantyPeriod;
        PowerConsumption = powerConsumption;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}; Гарантия - {WarrantyPeriod} года; Потребление энергии - {PowerConsumption} Вт";
    }

    public override void Discount()
    {
        Console.WriteLine($"Применена скидка на '{Name}' от производителя {Manufacturer} с гарантией {WarrantyPeriod} года. Итоговая стоимость - {Price * 0.75} руб., Гарантия переопределена - {WarrantyPeriod / 2} года");
    }

    public void CalculateEnergyCost(int hoursPerDay, int daysPerYear, double pricePerKWh)
    {
        double energyConsumption = PowerConsumption * hoursPerDay * daysPerYear / 1000.0;
        double totalCost = energyConsumption * pricePerKWh;
        Console.WriteLine($"Годовая стоимость энергии для {Name} - {totalCost} руб.\n");
    }
}

// Пример вызова метода GetInfo и вывода результата
Fiction diction = new Fiction("1984", "Издательство АСТ", 239, 0.5, "Дж. Оруэлл", 328, "Художественная литература");
Console.WriteLine(diction.GetInfo());
diction.CalculateReadingTime(20);

Clothing clother = new Clothing("Худи", "OffWhite", 19299, 0.8, 56, "Хлопок");
Console.WriteLine(clother.GetInfo());
clother.Display();
clother.CheckFit(56);

Electronics tv = new Electronics("Samsung G5120", "Samsung", 42999, 5.0, 4, 150);
Console.WriteLine(tv.GetInfo());
tv.Discount();
tv.Display();
tv.CalculateEnergyCost(5, 365, 0.1);

Books mybook = new Books("Удержать небо", "Издательство АСТ", 1099, 0.6, "Лю Цысинь", 448);
Console.WriteLine(mybook.GetInfo());
mybook.CalculateReadingTime(25);

// Generic-класс для хранения списка продуктов
public class ProductList<T> where T : Product
{
    private List<T> products = new List<T>();

    public void AddProduct(T product)
    {
        products.Add(product);
    }

    public void DisplayAll()
    {
        foreach (var product in products)
        {
            product.Display();
        }
    }
}

ProductList<Product> productList = new ProductList<Product>();
productList.AddProduct(diction);
productList.AddProduct(clother);
productList.AddProduct(tv);
productList.AddProduct(mybook);

productList.DisplayAll();
```
